```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data_ona/final_dataset.csv")
columns_to_use = list(df.columns[:18])
df_filtered = df[columns_to_use]
df_filtered.head()
```

binder chain - **A_seq** <br>
**binder** - whether it binds or not <br>
target chain - **B_seq** <br>

```python
Dict_targets = dict(df_filtered["target_id"].value_counts())
Dict_targets
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plt.figure(figsize=(6, 4))
sns.countplot(data = df_filtered, x = "target_id", ax=axes[0])
axes[0].tick_params(axis = "x", rotation = 90)
axes[0].set_title("Count by target_id")

plt.figure(figsize=(6, 4))
sns.countplot(data = df_filtered, x = "binder", ax=axes[1])
axes[1].tick_params(axis = "x", rotation = 90)
axes[0].set_title("Count by binder")

plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df_filtered, x="target_id", hue="binder")

plt.xticks(rotation=90)
plt.title("Counts of target_id grouped by binder")
plt.legend(title="Binder")
plt.tight_layout()
plt.show()
```
